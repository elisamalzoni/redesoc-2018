# Lacunas Estruturais

Importando bibliotecas:

```python
import pandas as pd
import networkx as nx
import socnet as sn

from random import shuffle
from itertools import combinations
```

Configurando bibliotecas:

```python
sn.node_size = 15

sn.edge_width = 1
sn.edge_color = (191, 191, 191)
```

Definindo constantes:

```python
FRAC_CLOSERS = 0.5

NUM_NODES = 25

DENSITY = 0.5

OPENER = 0
CLOSER = 1

OPENER_COLOR = (255, 0, 0)
CLOSER_COLOR = (0, 0, 255)
```

Exemplo de grafo:

(esse grafo representa o *ambiente* do modelo)

```python
def build_graph():
    g = nx.erdos_renyi_graph(NUM_NODES, DENSITY)

    sn.reset_node_colors(g)
    sn.reset_edge_colors(g)

    return g


g = build_graph()

sn.reset_positions(g)

sn.show_graph(g)
```

Exemplo de distribuio de tipos de n:

(esses tipos representam os *agentes* do modelo)

```python
def set_types(g, num_closers=0):
    nodes = list(g.nodes)
    shuffle(nodes)

    for n in nodes[num_closers:]:
        g.nodes[n]['type'] = OPENER

    for n in nodes[:num_closers]:
        g.nodes[n]['type'] = CLOSER

    for n in g.nodes:
        if g.nodes[n]['type'] == OPENER:
            g.nodes[n]['color'] = OPENER_COLOR
        else:
            g.nodes[n]['color'] = CLOSER_COLOR


set_types(g, int(NUM_NODES / 2))

sn.show_graph(g)
```

Custo de manter uma relao do n $n$:

$$c_n = \left\{\begin{array}{cl}
0             & \textrm{se }d_n = 0\textrm{;} \\
\frac{1}{d_n} & \textrm{caso contrrio,} \\
\end{array}\right.$$

onde $d_n$  o grau de $n$.

```python
def cost(g, n):
    dn = g.degree(n)

    if dn == 0:
        return 0
    return 1 / dn
```

Restrio do n $n$ para *openers* (frmula de Burt):

$$\sum_{m \in N_n} \left(c_n + \sum_{k \in N_n, k \in N_m} c_n c_k\right)^2\textrm{,}$$

onde $N_n$  o conjunto de vizinhos de $n$.

Esta frmula parte da premissa de que o n no est isolado.

Interpretao informal: o agente busca aumentar suas conexes e prefere conexes indiretas.

```python
def opener_constraint(g, n):
    constraint = 0

    for m in g.neighbors(n):
        sub_constraint = cost(g, n)

        for k in g.neighbors(n):
            if g.has_edge(k, m):
                sub_constraint += cost(g, n) * cost(g, k)

        constraint += sub_constraint**2

    return constraint
```

Restrio geral do n $n$:

* se o n estiver isolado,  igual a $2$;
* caso contrrio, depende de qual  o tipo.

```python
def constraint(g, n):
    if g.degree(n) == 0:
        return 2

    if g.nodes[n]['type'] == OPENER:
        return opener_constraint(g, n)
    return closer_constraint(g, n)
```

Inicializao da simulao:

```python
def initialize(g):
    for n in g.nodes:
        g.nodes[n]['constraint'] = constraint(g, n)
```

Passo da simulao:

1. escolha aleatoriamente um par de ns;
2. considere uma *inverso* da situao desse par, ou seja, remover a aresta se ela existir ou adicionar se no existir;
3. calcule o *ganho* de cada um desses ns, ou seja, a reduo da restrio como consequncia dessa inverso;
4. no caso particular em que a inverso foi adio, se um dos ganhos for negativo, os dois ganhos so redefinidos como zero *(justificativa no cdigo)*;
5. se um dos ganhos foi positivo aceite a inverso e termine o passo, seno volte para (1);
6. se todos os pares foram escolhidos e nenhuma inverso foi aceita, termine o passo.

```python
def invert(g, n, m):
    if g.has_edge(n, m):
        g.remove_edge(n, m)
    else:
        g.add_edge(n, m)
        g.edges[n, m]['color'] = sn.edge_color


def update(g):
    # Para cada par de ns, em ordem aleatria.

    pairs = list(combinations(g.nodes, 2))
    shuffle(pairs)

    for n, m in pairs:
        # Inverte a situao do par.
        invert(g, n, m)

        # Calcula as restries aps inverso.
        constraint_n = constraint(g, n)
        constraint_m = constraint(g, m)

        # Calcula os ganhos aps inverso,
        # (lembrando que restrio  ruim)
        gain_n = g.nodes[n]['constraint'] - constraint_n
        gain_m = g.nodes[m]['constraint'] - constraint_m

        # Se a inverso foi adio e um dos ns foi prejudicado,
        # ento o ganho  zero porque essa inverso  impossvel.
        # (voc no pode forar uma pessoa a manter uma relao)
        if g.has_edge(n, m) and (gain_n < 0 or gain_m < 0):
            gain_n = 0
            gain_m = 0

        # Se algum n teve um ganho, aceita a inverso e
        # considera o passo da simulao como terminado,
        # devolvendo True para indicar que houve mudana.
        if gain_n > 0 or gain_m > 0:
            g.nodes[n]['constraint'] = constraint_n
            g.nodes[m]['constraint'] = constraint_m
            return True

        # Se nenhum n teve um ganho, essa inverso no  aceita.
        # Restauramos, portanto, a situao original do par de ns.
        invert(g, n, m)

    # Se passamos por todos os pares e nenhuma inverso foi aceita,
    # considera o passo da simulao como terminado, devolvendo
    # False para indicar que no houve mudana: a rede estabilizou.
    return False
```

Simulao com visualizao, quando todos os ns buscam reduzir a restrio de Burt:

```python
def simulate(num_closers):
    g = build_graph()

    set_types(g, num_closers)

    initialize(g)

    while update(g):
        pass

    return g


num_closers = 0

g = simulate(num_closers)

sn.reset_positions(g)

sn.show_graph(g)
```

Inverso da visualizao:

```python

def inver_show(g):
    cg = nx.complement(g)

    sn.reset_positions(cg)

    for n in g.nodes:
        g.nodes[n]['pos'] = cg.nodes[n]['pos']

    sn.show_graph(g)


inver_show(g)
```

Distribuio de *betweenness*:

```python
betweenness = nx.betweenness_centrality(g)

df = pd.DataFrame({
    'betweenness': [betweenness[n] for n in g.nodes],
})

df.describe()
```

```python
num_closers = int(FRAC_CLOSERS * NUM_NODES)

g = simulate(num_closers)

sn.reset_positions(g)

sn.show_graph(g)
```
